**Import Lbirary**

```python
import pandas as pd
import numpy as np
```

**Import Dataset**

```python
df = pd.read_csv("https://github.com/sridharan-prof/Big-data-analyis/raw/main/netflix_data.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

**Get Feature Selection**

```python
df_features = df[['title','cast','rating','duration']].fillna('')
```

```python
df_features.shape
```

Selected four existing feautres to recommend movie.

```python
df_features
```

```python
x = df_features['title'] + ' ' + df_features['cast'] + ' ' + df_features['rating'] + ' ' + df_features['duration']
```

```python
x
```

```python
x.shape
```

**Get Feature Conversation to Tokens**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
x = tfidf.fit_transform(x)
```

```python
x.shape
```

```python
print(x)
```

**Get Similarity score using Cosine Similarity**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_Score = cosine_similarity(x)
```

```python
similarity_Score
```

```python
similarity_Score.shape
```

**Get Moive Name as Input from User and Validate for Closest Spelling**

```python
Favourite_movie_Name = input('Enter your favourite movie name: ')
```

```python
All_movies_title_List = df['title'].tolist()
```

```python
import difflib
```

```python
movie_recommendation = difflib.get_close_matches(Favourite_movie_Name, All_movies_title_List)
print(movie_recommendation)
```

```python
Close_match = movie_recommendation[0]
print(Close_match)
```

```python
index_of_close_match_movie = df[df.title == Close_match]['rating'].index[0]
print(index_of_close_match_movie)
```

```python
#getting a list of similar movies
Recomendation_score = list(enumerate(similarity_Score[index_of_close_match_movie]))
print(Recomendation_score)
```

```python
len(Recomendation_score)
```

**Get All Movies Sort Based on Recommendation Score with Facourite Movie**

```python
#sorting the movies based on the similarity scores
Sorted_similar_movies = sorted(Recomendation_score, key = lambda x:x[1], reverse = True)
print(Sorted_similar_movies)
```

```python
#print the name of the similar movies based on the index
print('Top 30 Movies suggested for you: \n')

i = 1
for movie in Sorted_similar_movies:
  index = movie[0]
  title_from_index = df[df.index == index]['title'].values[0]
  if (i<31):
    print(i, '.', title_from_index)
    i+=1
```

**Top 10 Movie Recommendation System**

```python
movie_name = input('Enter your favourite movie name: ')

list_of_all_titles = df['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = df[df.title == close_match]['rating'].index[0]

recommendation_score = list(enumerate(similarity_Score[index_of_the_movie]))

Sorted_similar_movies = sorted(recommendation_score, key = lambda x:x[1], reverse = True)

print('Top 10 Movies suggested for you: \n')

i = 1

for movie in Sorted_similar_movies:
  index = movie[0]
  title_from_index = df[df.index == index]['title'].values[0]
  if (i<11):
    print(i, '.', title_from_index)
    i+=1
```
